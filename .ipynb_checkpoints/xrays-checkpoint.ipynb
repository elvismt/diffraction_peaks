```python
import pandas as pd
from collections import OrderedDict
import os, json

data = OrderedDict()
for fname in os.listdir('.'):
    if fname.endswith('.json'):
        with open(fname, 'r') as f:
            js = json.load(f)
        data[js['name']] = pd.DataFrame({
            'x': js['x'], 'y': js['y']})
```

## The data

```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt

ncols = 2
nrows = math.ceil(len(data)/2)
fig, axs = plt.subplots(nrows, ncols, figsize=(15,8))
plt.subplots_adjust(hspace=0.4)
if len(data) != 0:
    fig.delaxes(axs[-1][-1])
for i, (k,v) in enumerate(data.items()):
    ax = axs[int(i/ncols)][int(i%ncols)]
    ax.set_title(k)
    ax.plot(v['x'], v['y'], 'c-')
```
